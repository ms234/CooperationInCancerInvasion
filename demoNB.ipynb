I've written my solver as a class, so to have all the parameters and options nicely packaged together. To use it we first create an instance.

```python
m = mySolver.fullModelSolver(domainSizeList=[0,1,75],dx=1e-2)
```

We have the following options when instantiating it:

Next, you initialise the system. You do so as follows:

```python
m.Set_ICParameters(stromaParamList=[0.2,0.1],tumourAParamList=[0.1,0.1],
               tumourMParamList=[0.1,0.1],matrixParamList=[0.2,0.1],
               acidParamList=[0.1,0.1])
m.ApplyInitialConditions()
```

```python
m.Plot([0])
```

Finally, to run it, we employ the Run() method.

```python
m.Run(rtol=1e-10,atol=1e-10)
```

And we can plot the results as follows:

```python
m.Plot([0,m.tEnd/2,m.tEnd],figsize=(8.27, 11.69),ylimVec=[-0.5,1.3])
```
